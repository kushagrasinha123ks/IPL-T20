# Economy
A bowler's economy rate, often referred to simply as "economy," is a key metric in cricket that measures the average number of runs conceded by the bowler per over bowled. It is a crucial indicator of a bowler's efficiency and effectiveness in restricting the scoring rate of the opposing batsmen.

Here's how the economy rate is calculated:

Economy Rate = (Total Runs Conceded / Total Overs Bowled)

```python
average_eco = df_bowlers['E/R'].mean()
most_economical = df_bowlers[df_bowlers['E/R']<=average_eco]
most_economical[['Player','E/R']]
```

```python
len(most_economical)
```

- Total 52 economical bowlers across all seasons.

# Most Wicket Taking Bowlers

```python
average_wicket = df_bowlers['Wkts'].mean()
most_wicket = df_bowlers[df_bowlers['Wkts']>=average_wicket]
most_wicket[['Player','Wkts']]
```

```python
len(most_wicket)
```

- 40 bowlers with high wicket taking capacity.

# Most Valuable bowler with more than average wickets and economy

```python
most_valuable = df_bowlers[(df_bowlers['Wkts']>average_wicket) & (df_bowlers['E/R']<average_eco)]
most_valuable
```

```python
len(most_valuable)
```

- 13 most valuable bowlers with lower economy and high wicket taking ability.

# Bowlers for Death-Overs
